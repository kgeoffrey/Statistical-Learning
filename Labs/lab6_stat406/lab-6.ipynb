```r
data  <-  read.table("Iris.txt", sep = ",", header = TRUE)
library(MASS)
```

```r
Tlda <- lda(y ~ sepal + petal, data=data)
```

```r
xvol <- seq(3, 8, length=200)
xrat <- seq(3, 8, length=200)
the.grid <- expand.grid(xvol, xrat)
names(the.grid) <- c('sepal', 'petal')
```

```r
pr.lda <- predict(Tlda, newdata=the.grid)$posterior
```

```r
image(xrat, xvol, matrix(pr.lda[, 2], 200, 200), col=terrain.colors(100), ylab='sepal', xlab='petal', main='LDA')
points(petal ~ sepal, data=data, pch=19, cex=1.5,col=c('red', 'blue')[y+1])
```

```r
contour("sepal", "petal", matrix(pr.lda, 200, 200),
        levels = .5, add = TRUE, drawlabels = FALSE, lwd=4, col='maroon')
```
